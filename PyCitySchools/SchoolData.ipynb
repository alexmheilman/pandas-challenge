Two Observable Trends Based on the Data: 
    Students showed higher achievement in reading than math throughout this analysis. This continued to be true even when other factors like school type, school budget, and spending on students was accounted for. 
    In reading and math, scores did not significatnly increase as students moved from 9th grade through 12th grade. The scores showed a consistent pattern of staying around the same average (ie. if it started at 80% in 9th grade, it was also shown to be 80% in 12th grade). 

```python
import pandas as pd
```

```python
# Import files to use 
school_data_load = 'schools_complete.csv'
student_data_load = 'students_complete.csv'

# Read in to DF
school_data = pd.read_csv(school_data_load)
student_data = pd.read_csv(student_data_load)

# Merge Data to one DF 
complete_data = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
complete_data.head()
```

# District Summary

```python
# Number of Unique schools
school_count = complete_data['school_name'].nunique()
school_count
```

```python
# Total Number of students 
student_count = complete_data['Student ID'].nunique()
student_count
```

```python
# Total Budget
complete_data['budget'].unique()
total_budget = 1910635 + 1884411 + 1056600 + 3022020 + 917500 + 1319574 + 1081356 + 3124928 + 248087 + 585858 + 1049400 + 2547363 + 3094650 + 1763916 + 1043130
print(total_budget)
```

```python
# Average Math Score 
average_math = ((complete_data['math_score'].sum())/student_count)
average_math
```

```python
# Average Reading Score
average_reading = ((complete_data['reading_score'].sum())/student_count)
average_reading
```

```python
# Percent of Students who Passed Math 
passing_math_count = complete_data[(complete_data["math_score"] >= 70)].count()["student_name"]
passing_math_percentage = passing_math_count / float(student_count) * 100
passing_math_percentage
```

```python
# Percent of Students who Passed Reading
passing_reading_count = complete_data[(complete_data["reading_score"] >= 70)].count()["student_name"]
passing_reading_percentage = passing_reading_count / float(student_count) * 100
passing_reading_percentage
```

```python
# Use the following to calculate the percentage of students that passed math and reading
passing_math_reading_count = complete_data[(complete_data["math_score"] >= 70) & (complete_data["reading_score"] >= 70)].count()["student_name"]
overall_passing_rate = passing_math_reading_count /  float(student_count) * 100
overall_passing_rate
```

```python
# Create a high-level snapshot of the district's key metrics in a DataFrame
district_summary = pd.DataFrame({'Total Schools': school_count, 'Total Students': student_count,'Total Budget': total_budget,'Average Math Score': average_math, 'Average Reading Score': average_reading, 'Students Passing Math (%)': passing_math_percentage,'Students Passing Reading (%)': passing_reading_percentage, 'Students Passing Reading and Math (%)': overall_passing_rate}, index = [0]) 

# Formatting
district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)

# Display the DataFrame
district_summary
```

# School Summary

```python
# Get Types of School
complete_data['type'].unique()
```

```python
# Calculate the total student count per type 
district_student_count = complete_data[(complete_data['type'] == 'District')].count()['student_name']
charter_student_count = complete_data[(complete_data['type'] == 'Charter')].count()['student_name']
print(district_student_count)
charter_student_count
```

```python
# Students Per School 
per_school_students = complete_data.groupby(['school_name']).count()['student_name']
per_school_students
```

```python
# Calculate the total school budget and per capita spending
per_school_budget = complete_data.groupby(["school_name"]).mean()["budget"]
per_school_capita = per_school_budget / district_student_count
print(per_school_capita)
per_school_budget
```

```python
# Calculate the average test scores
per_school_math = (complete_data.groupby(['school_name']).sum()['math_score'])/(complete_data.groupby(['school_name']).count()['student_name'])
per_school_reading = (complete_data.groupby(['school_name']).sum()['reading_score'])/(complete_data.groupby(['school_name']).count()['student_name'])
print(per_school_math)
per_school_reading
```

```python
# Calculate the number of schools with math scores of 70 or higher
school_passing_math = per_school_math.groupby(per_school_math >= 70).count()
school_passing_math
```

```python
# Calculate the number of schools with reading scores of 70 or higher
school_passing_reading = per_school_reading.groupby(per_school_reading >= 70).count()
school_passing_reading
```

```python
# Use the provided code to calculate the schools that passed both math and reading with scores of 70 or higher
passing_math_and_reading = (per_school_math.groupby(per_school_math >= 70).count()) & (per_school_reading.groupby(per_school_reading >= 70).count())
passing_math_and_reading
```

```python
# Use the provided code to calculate the passing rates
per_school_passing_math = complete_data.groupby(["school_name"]).count()["student_name"] / per_school_students * 100
per_school_passing_reading = complete_data.groupby(["school_name"]).count()["student_name"] / per_school_students * 100
overall_passing_rate = complete_data.groupby(["school_name"]).count()["student_name"] / per_school_students * 100
```

```python
# Create a DataFrame called `per_school_summary` with columns for the calculations above.
per_school_summary = pd.DataFrame({'District Student Count': district_student_count, 'Charter Student Count': charter_student_count, 'Students Per School': per_school_students, 'Budget Per School': per_school_budget, 'Budget Per Student': per_school_capita, 'Average Math Score Per School': per_school_math, 'Average Reading Score Per School': per_school_reading, 'Number of Schools Passing Math': school_passing_math, 'Number of Schools Passing Reading': school_passing_reading, 'Schools Passing Reading and Math': passing_math_and_reading, 'Math Passing Rate': per_school_passing_math, 'Reading Passing Rate': per_school_passing_reading, 'Overall Passing Rate': overall_passing_rate})

# Formatting
per_school_summary["Budget Per School"] = per_school_summary["Budget Per School"].map("${:,.2f}".format)
per_school_summary["Budget Per Student"] = per_school_summary["Budget Per Student"].map("${:,.2f}".format)

# Display the DataFrame
per_school_summary
```

# Highest Performing Schools

```python
highest_performing = per_school_summary.sort_values(by = ['Average Math Score Per School', 'Average Reading Score Per School'], ascending = [False, False])
highest_performing.head()
```

# Bottom Performing Schools 

```python
lowest_performing = per_school_summary.sort_values(by = ['Average Math Score Per School', 'Average Reading Score Per School'])
lowest_performing.head()
```

# Math Scores by Grade

```python
# Use the code provided to separate the data by grade
ninth_graders = complete_data[(complete_data["grade"] == "9th")]
tenth_graders = complete_data[(complete_data["grade"] == "10th")]
eleventh_graders = complete_data[(complete_data["grade"] == "11th")]
twelfth_graders = complete_data[(complete_data["grade"] == "12th")]
```

```python
# Combine to Single Data Frame 
math_scores_by_grade = pd.DataFrame({'Ninth Grade Scores': ninth_graders_scores, 'Tenth Grade Scores': tenth_graders_scores, 'Eleventh Graders Scores': eleventh_graders_scores, 'Twelfth Graders Scores': twelfth_graders_scores})
math_scores_by_grade.index.name = None
math_scores_by_grade
```

# Scores by School Spending

```python
# Establish the bins 
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
# Create a copy of the school summary since it has the "Per Student Budget" 
school_spending_df = per_school_summary.copy()
school_spending_df['Students Per School'] = school_spending_df['Students Per School'].astype(float)
school_spending_df['Budget Per School'] = school_spending_df['Budget Per School'].replace('[\,,]','',regex = True).replace('[\$,]','',regex = True).astype(float)
spending_per_students = school_spending_df['Budget Per School'] / school_spending_df['Students Per School']
spending_df = pd.DataFrame({'Spending Per Student': spending_per_students})
spending_df
```

```python
# Use `pd.cut` to categorize spending based on the bins.
spending_df["Spending Ranges (Per Student)"] = pd.cut(spending_df['Spending Per Student'], bins=spending_bins,labels = labels)
spending_df
```

# Scores by School Size

```python
# Establish the bins.
size_bins = [0, 1000, 2000, 5000]
labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
```

```python
# Categorize the spending based on the bins
# Use `pd.cut` on the "Total Students" column of the `per_school_summary` DataFrame.

per_school_summary["School Size"] = pd.cut(per_school_summary['Students Per School'], bins = size_bins, labels = labels)
per_school_summary
```

```python
# Calculate averages for the desired columns. 
size_math_scores = per_school_summary.groupby(["School Size"]).mean()["Average Math Score Per School"]
size_reading_scores = per_school_summary.groupby(["School Size"]).mean()["Average Reading Score Per School"]
size_passing_math = per_school_summary.groupby(["School Size"]).mean()["Math Passing Rate"]
size_passing_reading = per_school_summary.groupby(["School Size"]).mean()["Reading Passing Rate"]
size_overall_passing = per_school_summary.groupby(["School Size"]).mean()["Overall Passing Rate"]
```

```python
# Create a DataFrame called `size_summary` that breaks down school performance based on school size (small, medium, or large).
# Use the scores above to create a new DataFrame called `size_summary`
size_summary = pd.DataFrame({'Math Scores by Size': size_math_scores, 'Reading Scores by Size': size_reading_scores, 'Passing Math by Size': size_passing_math, 'Passing Reading by Size': size_passing_reading, 'Overall Passing by Size': size_overall_passing})

# Display results
size_summary
```
